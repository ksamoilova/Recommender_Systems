```python
# подсчет среднего рейтинга для каждого фильма

ratings_avg = ratings.groupby('movieId')['rating'].mean()
ratings_avg.head()
```

```python
# подстчет количества оценок для каждого фильма

num_rates = ratings.groupby('movieId')['rating'].count()
num_rates.head()
```

```python
# добавление к каждому фильму среднего рейтинга и количества оценок 
movies_ratings = movies.join (ratings_avg, on = 'movieId').rename(columns={'rating':'avg_rating'})
movies_rates = movies_ratings.join (num_rates, on = 'movieId').rename(columns={'rating':'rate_number'})

# новая метрика - произведение среднего рейтинга на количество оценок
movies_rates['movies_score'] = movies_rates['avg_rating']*movies_rates['rate_number']
```

```python
movies_rates.sort_values('movies_score', ascending=False).head()
```

### Постройте топ фильмов в категориях Action и Comedy

```python
# ТОП-10 фильмов в категориях Comedy

top10_comedies = movies_rates[movies_rates['genres'].str.contains('Comedy')].sort_values('movies_score', ascending=False)
top10_comedies.head(10)
```

```python
# ТОП-10 фильмов в категориях Action

top10_action = movies_rates[movies_rates['genres'].str.contains('Action')].sort_values('movies_score', ascending=False)
top10_action.head(10)
```
